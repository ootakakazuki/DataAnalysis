```python
# nullデータのあるレコードを削除
df = df.dropna()
df.shape
```

```python
# Dt_Customerカラムをdatetimeに直す

df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"])
```

```python
 df["Dt_Customer"] 
```

```python
print(df["Dt_Customer"][0])
```

```python
# 年月日のみを取得し、dates配列に入れる
dates = []
for i in df["Dt_Customer"]:
    dates.append(i.date())
print(max(dates))
print(min(dates))
```

```python

days = []
d1 = max(dates)
for date in dates:
    days.append(d1 - date) # 最新客の登録された年月日から各顧客の登録年月日を引いて、日数を算出している
df["customer_for"] = days
df["customer_for"] = pd.to_numeric(df["customer_for"], errors = "coerce")
#df["customer_for"][0]/(86400*365)
print(days[0])
print(d1)
print(dates[0])
print(days[1])
print(d1)
print(dates[1])
```

```python
# 特徴量を分析しやすいように変えていく
```

```python
# Year_birthカラムからAge(年齢)のカラムを作る　
this_year =  (datetime.datetime.now().date().strftime("%Y")) 
df["Age"] = int(this_year) - df["Year_Birth"]

# 総購入金額(単位はドル)
df["Spent"] = df["MntWines"] + df["MntFruits"] + df["MntMeatProducts"] + df["MntFishProducts"]+ df["MntSweetProducts"]+ df["MntGoldProds"]

# 誰かと住んでるならPartner, 1人ならAloneに変換する
df["LivingWith"] = df["Marital_Status"].replace({"Married": "Partner", "Together": "Partner", "single": "Alone","Divorced": "Alone", "Widow": "Alone","Absurd": "Alone","YOLO": "Alone", "Single": "Alone",})

# 同居している子供の数
df["Children"] = df["Kidhome"] + df["Teenhome"]

# 世帯人数
df["FamilySize"] = df["LivingWith"].replace({"Alone": 1, "Partner": 2}) + df["Children"]

# 親か否か 子供の数が1以上なら1、1人もいなければ0
df["IsParent"] = np.where(df.Children > 0, 1, 0) 


# 3つの教育レベルに分ける　Basic と2n Cycle => UnderGraduate,    Graduation =>  Graduate,    Master と PhD => PostGraduate
df["Education"] = df["Education"].replace({"Basic": "UnderGraduate", "2n Cycle": "UnderGraduate", "Graduation": "Graduate", "Master": "PostGraduate", "PhD": "PostGraduate"})

# 一部の列名をよりわかりやすく
df=df.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold"})

# 分析に不必要な列は削除する
to_drop = ["Marital_Status", "Dt_Customer", "Z_CostContact", "Z_Revenue", "Year_Birth", "ID"]
df = df.drop(to_drop, axis=1)

```

```python
# ここからはクラスタリング操作前の前処理を行います。

# カテゴリー変数にラベリング
# Standard Scalerで特徴量を標準化
# 次元削減用のデータセットを作る
```

```python
df.dtypes == "object"
```

```python
s = df.dtypes == "object"
```

```python
# object型のカラムを作る　ここではEducationとLivingWithが該当
s[s].index
```

```python
object_cols = list(s[s].index)
```

```python
# df2 作成(dfのコピー)

import copy
df2 = copy.deepcopy(df)
```

```python
# object型の変数をLabelEncoderでラベリングする
LE = LabelEncoder()
for i in object_cols:
    df2[i] = df[[i]].apply(LE.fit_transform) # なぜ[[i]]なのか調査中 ー＞　これでデータフレームになるようだ
```

```python
df2[object_cols]
```

```python
df2[["Education"]]
```

```python
df2.columns
```

```python
# 列を削除
cols_delite = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5','AcceptedCmp1', 'AcceptedCmp2','Complain', 'Response']
df2 = df2.drop(cols_delite, axis=1)
```

```python
scaler = StandardScaler()
scaler.fit(df2)
```

```python
scaled_df = pd.DataFrame(scaler.transform(df2), columns=df2.columns)
scaled_df
```

```python
# 次元削減

pca = PCA(n_components=3) # 3つの変数に絞り込む
#pca = PCA(n_components=2) # 2つの変数に絞り込む

pca.fit(scaled_df)

# データを主成分空間に写像
feature = pca.transform(scaled_df)

pca_df = pd.DataFrame(feature, columns=(["col1", "col2", "col3"]))
#pca_df = pd.DataFrame(feature, columns=(["col1", "col2"]))

pca_df.describe().T
```

```python
plt.figure(figsize=(4, 4))
plt.scatter(feature[:, 0], feature[:, 1], alpha=0.5, c=list(pca_df.iloc[:, 0]))
plt.grid()
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
# 寄与率　値が大きいほどデータの変動の大部分をその変数で説明できる　
# 第１主成分だけでデータの変動の29%を説明できる　
pca.explained_variance_ratio_
```

```python
# 累積寄与率の描画　# 第１主成分から第3主成分でおよそ52%を説明できる
import matplotlib.ticker as ticker
plt.gca().get_xaxis().set_major_locator(ticker.MaxNLocator(integer=True))

plt.plot([0] + list(np.cumsum(pca.explained_variance_ratio_)), "-o")
```

```python
"""
固有値　主成分分析においては、主成分の分散を指す。
固有値の値が1を切る主成分の場合は、結果への影響が少ないと考えられ分析する必要がないとされ、算出されたとしても分析には利用しないこともあります。

"""
# PCAの固有値
pca.explained_variance_


```

```python
# 凝集クラスタリングを使っていく
AC = AgglomerativeClustering(n_clusters=5)

yhat_AC = AC.fit_predict(pca_df)
pca_df["Clusters"] = yhat_AC
pca_df["Clusters"]

# 元のデータにも入れる
df["Clusters"] = yhat_AC
```
