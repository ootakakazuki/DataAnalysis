```python
# SVC サポートベクターマシンのとき
from sklearn.svm import SVC
svm = SVC(random_state=1)
svm.fit(x_train.T, y_train.T)

acc = svm.score(x_test.T, y_test.T) * 100
accuracies["SVM"] = acc

print("サポートベクターマシンの正確さは{:.2f}%".format(acc))
```

```python
# ナイーブベイズ

from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train.T, y_train.T)

acc = nb.score(x_test.T, y_test.T) * 100
accuracies["NaiveBayes"] = acc
print("ナイーブベイズの正確さは{:.2f}%".format(acc))
```

```python
# 決定木　
# データから木構造の予測モデルを作る教師あり学習の手法
# 決定木単体では過学習に陥りやすいため、木の深度を制限したりする必要がある

from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train.T, y_train.T)

acc = dtc.score(x_test.T, y_test.T) * 100

accuracies["Decision Tree"] = acc
print("決定木の正確さは{:.2f}%".format(acc))
```

```python
# ランダムフォレスト
# ランダムフォレストはバギングを使用
# n_estimators : 決定木をいくつ用意するか 多いほど正確になるが、その反面、すべて学習するまで時間がかかる

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000, random_state=2)
rf.fit(x_train.T, y_train.T)

acc = rf.score(x_test.T, y_test.T) * 100
accuracies["Random Forest"] = acc
print("ランダムフォレストの正確さは{:.2f}%".format(acc))
```
